Normalization using Numpy

```python
# 12 , 22, 35, 56, 78, 89
#   6 , 5, 4, 3, 2, 1
```

```python
import numpy as np
data = np.array([1,2,39,400,5])
```

```python
min_val = np.min(data)
max_val = np.max(data)
print(min_val)
print(max_val)
```

```python
normalized_arr = (data-min_val)/(max_val-min_val)
normalized_arr
```

Explore About Z-Score Normalization

```python
data
```

```python
mean_val = np.mean(data)
std_val = np.std(data)
```

```python
normalize_val= (data-mean_val)/std_val
```

```python
normalize_val
```
