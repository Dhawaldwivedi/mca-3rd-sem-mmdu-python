```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Box plot using matplotlib
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(6, 4))
plt.boxplot(data, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.show()
```

```python
# Pie chart using matplotlib
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')
plt.show()
```

```python
# scatter plot
import numpy as np
x= np.random.randint(1,50,20)
y = np.random.randint(1,50,20)
plt.figure(figsize=(3,4))
plt.scatter(x,y,alpha= 0.9)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("scatter plot")
plt.show()
```

```python
# plt.hist (histogram)
x= np.random.rand(50)
plt.figure(figsize=(3,4),edgecolor='white')
plt.hist(x,bins=10)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("histogram plot")
plt.show()
```

```python
# seaborn library
import seaborn as sns
```

```python
# Histogram using seaborn
tips = sns.load_dataset("tips")
df = tips.head(10)
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x="total_bill", bins=20, kde=True)
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot using seaborn
plt.figure(figsize=(6, 4))
sns.barplot(x="day", y="total_bill", data=tips, ci=None)
plt.title('Average Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill')
plt.show()
```

```python
# Count plot using seaborn
plt.figure(figsize=(6, 4))
sns.countplot(x="day", data=tips)
plt.title('Count of Observations by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()
```

```python
# Boxen plot using seaborn
plt.figure(figsize=(6, 4))
sns.boxenplot(x="day", y="total_bill", data=tips)
plt.title('Boxen Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Swarm plot using seaborn
plt.figure(figsize=(6, 4))
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title('Swarm Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Heatmap using seaborn (correlation matrix of numerical columns)
plt.figure(figsize=(6, 4))
correlation_matrix = tips.select_dtypes(include=np.number).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
